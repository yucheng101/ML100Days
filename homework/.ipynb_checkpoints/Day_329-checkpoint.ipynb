### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## Q1
#  提示: 用 cdf 來算
mu=0
sigma=1
x=np.linspace(-4, 4, 500)

cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
```

```python
cumsum_probs[x<1][-1]
```

```python
k = stats.norm.ppf(cumsum_probs ,mu, sigma)
k
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
(1-cumsum_probs[x>1][0])+cumsum_probs[x<-1][-1]
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python
## Q3
#  提示: 和 Q1,Q2 不同，要先標準化。
mu=2
sigma=4
x=np.linspace(-12, 16, 500)

cumsum_probs =  stats.norm.pdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' pdf of Nomal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
```

```python
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
cumsum_probs[x<6][-1]
```
