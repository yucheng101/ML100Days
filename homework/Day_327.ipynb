### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
p = 0.5 # 事件A 機率 0.4
n = 100  # 重複實驗5次,
r = np.arange(0,101) # 可以出現的範圍為 0,1,2,...,5-->6種可能出現的結果
#print(type(k))

# 2.計算二項分佈的概率質量分佈 (probability mass function)
# 之所以稱為質量，是因為離散的點
# P(X=x) --> 是機率
probs = stats.binom.pmf(r, n, p)
print(probs)
#array([ 0.07776, 0.2592 , 0.3456 , 0.2304 , 0.0768 , 0.01024])
plt.bar(r, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('binomial(n=100,p=0.5)')
plt.show()
```

```python
cumsum_probs = stats.binom.cdf(r, n, p)
plt.show()
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('binomial(n=5,p=0.4)')
plt.plot(r, cumsum_probs)
plt.show()
```

```python
p_loc= stats.binom.ppf(cumsum_probs, n, p)
print(p_loc)
```

```python
X = stats.binom.rvs(n,p,size=5)
print(X)
plt.hist(X)
plt.show()
```
