### Q2: 運用範例的 python 程式碼，計算大樂透的中頭獎機率?

```python
N=49
K=6
n=6

#產生一個同樣間隔的序列
r = np.arange(0, 7) #產出 x 對應點
print(r)

# 2.計算機率質量函數 (probability mass function)
# P(X=x) --> 是機率
probs = stats.hypergeom.pmf(r, N,K,n)
print(probs)
print(type(probs))
print(1/probs)
#print(1/probs[4])
#print(1/probs[5])

plt.bar(r, probs)
plt.ylabel('P(X=x)')
plt.xlabel('x')
plt.title('pmf of Hypergeometric(N=49,K=6,n=6)')
plt.show()
```

```python
cumsum_probs = stats.hypergeom.cdf(r, N,K,n)
plt.show()
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title('cdf of Hypergeometric(N=50,K=30,n=10)')
plt.plot(r, cumsum_probs)
plt.show()
```
